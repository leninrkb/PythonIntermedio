```python
import random 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
n = [random.randint(0, 10) for _ in range(10000)]
x, frequency = np.unique(n, return_counts=True)
print(x, frequency)
sns.set_style('whitegrid')
axes = sns.barplot(x=x, y=frequency)

# customize
axes.set_title('random numbers')
axes.set_xlabel('numbers')
axes.set_ylabel('frequency')
axes.set_ylim(top=max(frequency)*1.2)
for bar, f in zip(axes.patches, frequency):
    text_x_position = bar.get_x()
    text_y_position = bar.get_height()+0.5
    percentage = int(f / len(n) * 1000) / 10
    text = f'{f}\n{percentage}%'
    axes.text(x=text_x_position, y=text_y_position, s=text)
```
